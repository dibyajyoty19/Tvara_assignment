```python
model = SentenceTransformers("infloat/e5-small-v2")
sentences = [" passage: I live in India",
             "passage: India is the most progressive country",
             "passage: AI is transforming the world"]

sentence_embeddings = model.encode(sentences, normalize_embeddings = True)
query = "query: Am indian"
query_embeddings = model.encode(query, normalize_embeddings=True)

scores = util.cos_sim(query_embeddings, sentence_embeddings)
best = scores.argmax()
print(sentences[best])
```
